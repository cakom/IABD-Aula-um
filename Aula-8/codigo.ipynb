```python
#Aula 8 - Exemplo de Arvores de DecisÃ£o

#importar biblioteca 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split #treino e teste
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree


```

```python
! pip install seaborn
```

```python
#criando  o conjundo  de dados 
notas = np.array([8,6,5,9,4,3,7,2,1,5,6,9,8])

classes = np.array(['bom','ruim','ruim','bom','pessimo','pessimo','bom','pessimo','pessimo','ruim','ruim','bom','bom'])
```

```python
#dividindo o conjunto de dados em treino e teste
notas_treinos, notas_teste, classes_treino, classes_teste = train_test_split(
    notas.reshape(-1,1), classes.reshape(-1,1), test_size=0.2, random_state=42)

    #80% treino e 20% teste 
```

```python
modelo = DecisionTreeClassifier() #criando o modelo
modelo.fit(notas_treinos, classes_treino) #treinando o modelo
```

```python
print ("notas teste", notas_teste  )
print ("classes teste", classes_teste  )
```

```python
#realizando a prediÃ§Ã£o 
modelo.predict(notas_teste)
```

```python
#realizando prediÃ§Ã£o com uma nova nota
modelo.predict(np.array([[6]])) # realiza prediÃ§Ãµes de notas individuais
```

```python
modelo.predict(np.array([7.5]).reshape (-1,1)) # realiza prediÃ§Ãµes de notas individuais
```

```python
fig = plt.figure(figsize=(10,12))# cria uma figura 10x12
tree.plot_tree(modelo,feature_names=notas.tolist(), class_names=classes.tolist(), filled=True) #     plota a Ã¡rvore de decisÃ£o
plt.title ("Ãrvore de DecisÃ£o -para classifiar bebidas de acordo com nota")
fontsize=16
plt.show()
```

```python
plt.figure(figsize=(10,8))# cria uma figura 10x12
plt.title('box plot das notas  ')
sns.boxplot(x=classes,y=notas,palette=['green','orange','red'])
```

```python
#grafico de barras
plt.figure(figsize=(10,8))
sns.barplot(y=notas,x=classes,palette=['green','yellow','red'])
plt.title('Notas das bebidas',fontsize=16)
plt.xlabel('classificaÃ§Ã£o',fontsize=16)
plt.ylabel('notas',fontsize=16)
```
